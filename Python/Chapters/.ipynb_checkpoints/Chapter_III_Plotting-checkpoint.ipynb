#### plt.subplots 

`fig, (ax1,ax2) =plt.subplots(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)`
Create a figure and a set of subplots. **Т.е. просто позволяет создать разыне графики на одной картинке.**
- `sharex, sharey : bool or {'none', 'all', 'row', 'col'}, default: False`

`fig` (можно самому выбрать имя) - создает пространство для графика (т.е. оси без заполнения)  
`ax1, ax2` - заполнение места под ваш граф 1 и 2

```python
#First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

#Creates two subplots and unpacks the output array immediately
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

**Полезный кусочек кода**

```python
# определяем размеры subplot
num_rows, num_cols = 3, 3
# какие признаки (колонки) будем строить
features = df[df[corr_list] + df['Platform']]
# объявляем subplot
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(10, 6))
# через enumerate получаем позиции: ряд = idx/num_cols, позиция в ряду = idx%num_cols 
for idx, feat in  enumerate(features):
    sns.boxplot(x='Platform', y=feat, data=df[df[corr_list] + df['Platform']], ax=axes[idx / num_cols, idx % num_cols])
    axes[idx / num_cols, idx % num_cols].legend()
    axes[idx / num_cols, idx % num_cols].set_xlabel('Platform')
    axes[idx / num_cols, idx % num_cols].set_ylabel(feat);
```

## Plotly (интерактивные графики)

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

Для построения интеракитивного графика надо подготовить массивы для линий и стиль оформления:

```python
# посчитаем число вышедших игр и проданных копий по годам/ df1.join(df2)
years_df = df.groupby('Year_of_Release')[['Global_Sales']].sum().join(
                      df.groupby('Year_of_Release')[['Name']].count())

years_df.columns = ['Global_Sales', 'Number_of_Games']
years_df.head(5)
```

```python
# создаем линию для числа проданных копий
trace0 = go.Scatter(
    x=years_df.index,
    y=years_df.Global_Sales,
    name='Global Sales'
)
trace0
```

```python
# создаем линию для числа вышедших игр 
trace1 = go.Scatter(
    x=years_df.index,
    y=years_df.Number_of_Games,
    name='Number of games released'
    )
trace1
```

```python
# определяем массив данных и задаем title графика в layout
data = [trace0, trace1]
layout = {'title': 'Statistics of video games'}
```

```python
# cоздаем объект Figure и визуализируем его
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

В plotly строится визуализация объекта Figure, который состоит из данных (массив линий, которые в библиотеке называются traces) и оформления/стиля, за который отвечает объект layout. `iplot` строит график от dict-like object т.е. {'name':'plot_name', 'x' : x_array, 'y': y_array}.

Параметр show_link отвечает за ссылки на online-платформу plot.ly на графиках. Поскольку обычно это функциональность не нужна, то я предпочитаю скрывать ее для предотвращения случайных нажатий.

```python
iplot(go.Figure(trace0))  
```

```python
# сохранить график в виде html-файла.
plotly.offline.plot(fig, filename='years_stats.html', show_link=False)
```

#### Bar plot

```python
platforms_df2 = df.groupby('Platform')[['Global_Sales']].sum()
platforms_df2.head(5)
```

```python
# считаем число проданных и вышедших игр по платформам
platforms_df = df.groupby('Platform')[['Global_Sales']].sum().join(
    df.groupby('Platform')[['Name']].count()
)
platforms_df.columns = ['Global_Sales', 'Number_of_Games']
platforms_df.sort_values('Global_Sales', ascending=False, inplace=True)
platforms_df.head(5)

#pd1.join(pd2) - использует общие индексы (row). В данном случаии индексы - это названия платформ
```

```python
# создаем traces для визуализации
trace0 = go.Bar(
    x=platforms_df.index,
    y=platforms_df.Global_Sales,
    name='Global Sales'
)

trace1 = go.Bar(
    x=platforms_df.index,
    y=platforms_df.Number_of_Games,
    name='Number of games released'
)

# создаем массив с данными и задаем title для графика и оси x в layout
data = [trace0, trace1]
layout = {'title': 'Share of platforms', 'xaxis': {'title': 'platform'}}

# создаем объект Figure и визуализируем его
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

#### Box plot

```python
# создаем Box trace для каждого жанра из наших данных
data = []
for genre in df.Genre.unique():
# df.Genre.unique()-вернет list с жантрами -> цикл по жанрам. data = [box1, box2, ...], 
# где box# - {'name':'strategy', 'y': array(Critic_Score for all strategies))}. 
# Итого data состоит из trace# соответствующим данному жанру
    data.append(
        go.Box(y=df[df.Genre==genre].Critic_Score, name=genre)
    )

# визуализация все Box traces
iplot(data, show_link = False)
```
